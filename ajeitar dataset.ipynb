```python
import pandas as pd
```

```python
df1 = pd.read_csv("spotify_millsongdata.csv")
df2 = pd.read_csv("tcc_ceds_music.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
dic = {df2.artist_name[n]:df2.genre[n] for n in range(len(df2))}
```

```python
df1["artist"] = df1["artist"].str.lower()
df1['genre'] = df1['artist'].map(dic)
df1.dropna(subset=['genre'], inplace=True)
df1.drop('link', axis=1, inplace=True)
```

```python
df1
```

```python
df1.rename(columns={"song": "track_name"}, inplace=True)
df1.rename(columns={"text": "lyrics"}, inplace=True)
```

```python
df1
```

```python
df1.to_csv('dataset_atualizado.csv', index=False)
```
