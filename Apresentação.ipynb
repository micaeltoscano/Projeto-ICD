## 1 - Introdução

A música exerce um papel significativo na vida cotidiana, influenciando nossas emoções e estado mental de maneiras profundas. Este projeto busca explorar a relação entre música e saúde mental, utilizando dados para investigar como diferentes gêneros musicais e hábitos musicais podem impactar o bem-estar psicológico. Através da análise de padrões e correlações, nosso objetivo é identificar possíveis efeitos positivos e negativos da música na saúde mental. Assim, este estudo contribui para uma compreensão mais profunda dos benefícios da música para o equilíbrio emocional.

### 1.1 - Objetivo geral

Este projeto tem como intuito investigar a influência de diferentes gêneros musicais e hábitos de escuta no bem-estar mental, buscando identificar padrões de correlação entre música e sintomas de saúde mental, como ansiedade e depressão.(opção 1)

A crescente preocupação com a saúde mental, especialmente em um mundo mais acelerado, conectado, com desafios sociais e emocionais complexos, reforça a necessidade de explorar abordagens alternativas e complementares para o bem-estar psicológico. ### 3.1 - Quais os gêneros favoritos de quem tem depressão, ansiedade, toc e insônia?

```python
df_generos = df["genero_fav"].unique()

ansiedade = df[df["ansiedade"] > 4]
depressao = df[df["depressao"] > 4]
toc = df[df["toc"] > 4]
insonia = df[df["insonia"] > 4]

contagem_ansiedade = ansiedade['genero_fav'].value_counts().reindex(df_generos, fill_value=0)
contagem_depressao = depressao['genero_fav'].value_counts().reindex(df_generos, fill_value=0)
contagem_insonia = insonia['genero_fav'].value_counts().reindex(df_generos, fill_value=0)
contagem_toc = toc['genero_fav'].value_counts().reindex(df_generos, fill_value=0)

df_efeitos = pd.DataFrame({
    'Genero': df_generos,
    'Ansiedade': contagem_ansiedade.values,
    'Depressão': contagem_depressao.values,
    'Insônia': contagem_insonia.values,
    'TOC': contagem_toc.values
})
```

### 3.5 - Como a música ajuda pessoas de diferentes faixas etárias a lidarem com transtornos mentais?

```python
idade = {}

melhora = df[df["efeitos_na_saude_mental"] == "Improve"]

generos = df["genero_fav"].unique() 

for gen in generos:
    idade[gen] = {} 
    
    for start in range(0, 100, 20):
        end = start + 20
        faixa_etaria = f"{start}-{end}"
        
        quant_por_faixa = melhora[(melhora["idade"] > start) & (melhora["idade"] <= end) & (melhora["genero_fav"] == gen)]
        
        idade[gen][faixa_etaria] = len(quant_por_faixa)
```

```python
df_idade = pd.DataFrame(idade)
df_idade.columns.name = 'Faixa etaria'
df_idade_T = df_idade.T
```

```python
plt.figure(figsize=(12, 6))
df_idade_T.plot(kind='bar', stacked=False, alpha=0.75)

plt.title('Distribuição de Gêneros Musicais por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Ouvintes')
plt.xticks(rotation=45)
plt.legend(title='Gênero Musical', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

### 3.9 - Quantas pessoas de determinadas faixas etárias costumam ouvir os gêneros?

```python
generos = list(set(['classica', 'country', 'edm',
       'folk', 'gospel', 'hip_hop',
       'jazz',  'metal', 'pop',
       'rb', 'rap', 'rock']))

idade = {gen: {} for gen in generos}

for gen in generos:
    nome_genero = gen.split('[')[-1].replace(']', '')  
    ouvintes_frequentes = df[df[gen] > 1]  
    
    for inicio in range(0, 100, 20):
        fim = inicio + 20
        faixa_etaria = f"{inicio}-{fim}"
        
        ouvintes_por_idade = ouvintes_frequentes[(ouvintes_frequentes["idade"] > inicio) & (ouvintes_frequentes["idade"] <= fim)]
        
        idade[gen][faixa_etaria] = len(ouvintes_por_idade)
```

```python
df_idade_genero2 = pd.DataFrame(idade).T
df_idade_genero2.columns = [col.split('[')[-1].replace(']', '') for col in df_idade_genero2.columns]
```

```python
plt.figure(figsize=(12, 6))
df_idade_genero2.plot(kind='bar', stacked=False, alpha=0.75)

# Adicionando título e rótulos
plt.title('Distribuição de Ouvintes por Faixa Etária e Gênero Musical')
plt.xlabel('Faixa Etária')
plt.ylabel('Número de Ouvintes')
plt.xticks(rotation=45)  # Rotaciona os rótulos do eixo x
plt.legend(title='Gênero Musical')

# Exibe o gráfico
plt.show()
```
