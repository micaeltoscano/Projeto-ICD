### 3.1 - Como a música se comporta ao longo das idades?

#### 3.1.1 - Quais os gêneros favoritos por faixa etária?

Nesta subseção, investigamos a relação entre a idade e os gêneros musicais preferidos dos ouvintes. Utilizamos um gráfico de barras para visualizar o número de ouvintes por faixa etária e gênero, permitindo observar tendências claras de preferência musical ao longo das diferentes fases da vida.

```python
generos = list(set(['classica', 'country', 'edm',
       'folk', 'gospel', 'hip_hop',
       'jazz',  'metal', 'pop',
       'rb', 'rap', 'rock']))

idade = {gen: {} for gen in generos}

for gen in generos:
    ouvintes_frequentes = df[df[gen] > 1]  
    
    for inicio in range(0, 80, 20):
        fim = inicio + 20
        faixa_etaria = f"{inicio}-{fim}"
        
        ouvintes_por_idade = ouvintes_frequentes[(ouvintes_frequentes["idade"] > inicio) & (ouvintes_frequentes["idade"] <= fim)]
        
        idade[gen][faixa_etaria] = len(ouvintes_por_idade)
```

```python
df_idade_genero2 = pd.DataFrame(idade).T
```

```python
plt.figure(figsize=(12, 6))
df_idade_genero2.plot(kind='bar', stacked=False, alpha=0.75)

plt.title('Distribuição de Ouvintes por Faixa Etária e Gênero Musical')
plt.xlabel('Gêneros')
plt.ylabel('Número de Ouvintes')
plt.xticks(rotation=45)  
plt.legend(title='Faixa etária')

plt.show()
```

Desse modo, concluímos que:
- entre 0-20 anos: Pop é o gênero favorito
  
- entre 20-40 anos: Rock é o gênero favorito

- entre 40-60 anos: Rock é o gênero favorito

- entre 60-80 anos: Música clássica é o gênero favorito

#### 3.1.2 - Quais os gêneros que mais ajudam as pessoas de diferentes faixas etárias a lidarem com transtornos mentais?

Aqui, analisamos quais gêneros são mais eficazes para melhorar a saúde mental de pessoas em diferentes faixas etárias. Utilizamos novamente gráficos de barras para mostrar a distribuição dos gêneros que ajudam a melhorar as condições de saúde mental em cada grupo etário.

```python
idade = {}

melhora = df[df["efeitos_na_saude_mental"] == "Improve"]

generos = df["genero_fav"].unique() 

for gen in generos:
    idade[gen] = {} 
    
    for start in range(0, 80, 20):
        end = start + 20
        faixa_etaria = f"{start}-{end}"
        
        quant_por_faixa = melhora[(melhora["idade"] > start) & (melhora["idade"] <= end) & (melhora["genero_fav"] == gen)]
        
        idade[gen][faixa_etaria] = len(quant_por_faixa)
```

```python
df_idade = pd.DataFrame(idade)
df_idade.columns.name = 'Faixa etaria'
df_idade_T = df_idade.T
```

```python
plt.figure(figsize=(12, 6))
df_idade_T.plot(kind='bar', stacked=False, alpha=0.75)

plt.title('Distribuição de Gêneros Musicais por Faixa Etária')
plt.xlabel('Gênero')
plt.ylabel('Número de Ouvintes')
plt.xticks(rotation=45)
plt.legend(title='Faixa etária', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

Desse modo, concluímos que entre todas as idades, o rock é o gênero que mais demonstrou ajudar as pessoas com suas condições mentais

#### 3.1.5 - Quais as possíveis análises que podem ser feitas entre a idade e o BPM das músicas?

Utilizando técnicas de clusterização (KMeans), examinamos a relação entre idade e o ritmo das músicas (BPM), destacando grupos etários com preferências distintas por músicas mais rápidas ou lentas.

```python
X = df[['idade', 'bpm']]
X = X.dropna(subset=['idade', 'bpm']) 

inertias = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

#mostrando o "cotovelo" do gráfico.

sns.lineplot(x=range(1,11), y=inertias) 
```

Criamos o gráfico para analisar um número ideal de clusters, que, nesse caso, foi 4.

```python
kmeans = KMeans(n_clusters=4, random_state=62) 
kmeans.fit(X) 
```

```python
X['cluster'] = kmeans.labels_           
sns.scatterplot(data=X, x='bpm', y='idade', hue='cluster', palette='bright')
```

Com a exibição desse gráfico, podemos analisar que existe 4 grupos, que podem ser divididos em 2. Os grupos [azul,verde,laranja] mostram que as pessoas entre 10 a 40 anos, apresentam uma maior diversidade em relação a velocidade da música. O grupo [vermelho] mostra que pessoas mais velhas, tendem a ter uma preferência por músicas mais calmas.

### 3.2 - Qual o impacto da música na saúde mental?

#### 3.2.1 - Existe uma relação entre o BPM e algum transtorno mental?

- Seguindo a lógica do gráfico anterior, será que conseguimos achar alguma relação entre pessoas com ansiedade com os grupos?

```python
X['cluster'].value_counts()
```

Contando Quantas Pessoas dos Clusters tem Ansiedade Acima de 4:

```python
X['cluster'] == 0
X['ansiedade'] = df['ansiedade']

pessoas_com_ansiedade_cluster0 = X[(X['cluster'] == 0) & (X['ansiedade'] > 4)]
pessoas_com_ansiedade_cluster1 = X[(X['cluster'] == 1) & (X['ansiedade'] > 4)]
pessoas_com_ansiedade_cluster2 = X[(X['cluster'] == 2) & (X['ansiedade'] > 4)]
pessoas_com_ansiedade_cluster3 = X[(X['cluster'] == 3) & (X['ansiedade'] > 4)]
pessoas_com_ansiedade = X[X['ansiedade'] > 4]
```

Calculando a Porcentagem de Pessoas com Ansiedade Acima de 4 no Cluster 0, 1, 2, 3:

```python
porcentagem_ansiedade_cluster0 = (len(pessoas_com_ansiedade_cluster0) / len(X[X['cluster'] == 0])) * 100
porcentagem_ansiedade_cluster1 = (len(pessoas_com_ansiedade_cluster1) / len(X[X['cluster'] == 1])) * 100
porcentagem_ansiedade_cluster2 = (len(pessoas_com_ansiedade_cluster2) / len(X[X['cluster'] == 2])) * 100
porcentagem_ansiedade_cluster3 = (len(pessoas_com_ansiedade_cluster3) / len(X[X['cluster'] == 3])) * 100

print(f"{porcentagem_ansiedade_cluster0:.2f}% de pessoas com ansiedade acima de 4 no grupo azul")
print(f"{porcentagem_ansiedade_cluster1:.2f}% de pessoas com ansiedade acima de 4 no grupo laranja")
print(f"{porcentagem_ansiedade_cluster2:.2f}% de pessoas com ansiedade acima de 4 no grupo verde")
print(f"{porcentagem_ansiedade_cluster3:.2f}% de pessoas com ansiedade acima de 4 no grupo vermelho")
```

Concluimos, portanto, que, das 175 que participam do grupo azul, 114 delas tem ansiedade maior do que nível 4. Além disso, nessa análise, vemos que mais de 73% das pessoas que escutam música agitada são ansiosas.

#### 3.2.2 - Quais os gêneros favoritos de quem tem transtornos mentais?

Investigamos quais gêneros musicais são mais populares entre pessoas com níveis significativos de ansiedade, depressão, insônia e TOC. Um gráfico de barras mostra a distribuição de preferências musicais para cada condição psicológica.

```python
df_generos = df["genero_fav"].unique()

ansiedade = df[df["ansiedade"] > 4]
depressao = df[df["depressao"] > 4]
toc = df[df["toc"] > 4]
insonia = df[df["insonia"] > 4]

contagem_ansiedade = ansiedade['genero_fav'].value_counts().reindex(df_generos, fill_value=0)
contagem_depressao = depressao['genero_fav'].value_counts().reindex(df_generos, fill_value=0)
contagem_insonia = insonia['genero_fav'].value_counts().reindex(df_generos, fill_value=0)
contagem_toc = toc['genero_fav'].value_counts().reindex(df_generos, fill_value=0)

df_efeitos = pd.DataFrame({
    'Genero': df_generos,
    'Ansiedade': contagem_ansiedade.values,
    'Depressão': contagem_depressao.values,
    'Insônia': contagem_insonia.values,
    'TOC': contagem_toc.values
})
```

```python
df_efeitos.set_index('Genero', inplace=True)

plt.figure(figsize=(12, 6))
df_efeitos.plot(kind='bar', stacked=False, alpha=0.75)

plt.title('Distribuição de Gêneros Musicais Favoritos por Condição Psicológica')
plt.xlabel('Gênero Musical')
plt.ylabel('Contagem de Ouvintes')
plt.xticks(rotation=45)
plt.legend(title='Condições Psicológicas')

plt.tight_layout()  
plt.show()
```

Observamos que o rock é o gênero favorito entre as pessoas que tem ansiedade, depressão, insônia e TOC, enquanto gospel é o menos favorito

Nesta subseção, comparamos os níveis médios de sintomas de transtornos mentais (ansiedade, depressão, insônia, TOC) para diferentes gêneros musicais. A análise foi representada por um gráfico de barras que ilustra o impacto dos gêneros nas condições psicológicas.

```python
sns.set(style="whitegrid")

# Reestruturando os dados: média dos sintomas de saúde mental para diferentes níveis de frequência dos gêneros
df_generos = df[['classica', 'country', 'edm', 'folk', 'gospel', 'hip_hop', 'jazz',  
            'metal', 'pop', 'rb', 'rap', 'rock']]

# Calculando a média dos sintomas de saúde mental para cada gênero musical
df_efeitos = pd.DataFrame({
    'Gênero': ['Rock', 'Pop', 'Hip-Hop', 'Música Clássica', 'Rap'],
    'Depressão': [
        df.loc[df['rock'] > 0, 'depressao'].mean(),
        df.loc[df['pop'] > 0, 'depressao'].mean(),
        df.loc[df['hip_hop'] > 0, 'depressao'].mean(),
        df.loc[df['classica'] > 0, 'depressao'].mean(),
        df.loc[df['rap'] > 0, 'depressao'].mean()
    ],
    'Insônia': [
        df.loc[df['rock'] > 0, 'insonia'].mean(),
        df.loc[df['pop'] > 0, 'insonia'].mean(),
        df.loc[df['hip_hop'] > 0, 'insonia'].mean(),
        df.loc[df['classica'] > 0, 'insonia'].mean(),
        df.loc[df['rap'] > 0, 'insonia'].mean()
    ],
    'TOC': [
        df.loc[df['rock'] > 0, 'toc'].mean(),
        df.loc[df['pop'] > 0, 'toc'].mean(),
        df.loc[df['hip_hop'] > 0, 'toc'].mean(),
        df.loc[df['classica'] > 0, 'toc'].mean(),
        df.loc[df['rap'] > 0, 'toc'].mean()
    ]
})

# Reorganizando os dados para formato longo (necessário para o gráfico de barras)
df_melted = df_efeitos.melt(id_vars='Gênero', var_name='Sintoma', value_name='Nível Médio')

# Criando o gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='Gênero', y='Nível Médio', hue='Sintoma', data=df_melted, palette='viridis')

# Configurando o gráfico
plt.title('Efeito dos Gêneros Musicais nos Níveis de Sintomas de Saúde Mental')
plt.xlabel('Gênero Musical')
plt.ylabel('Nível Médio dos Sintomas de Saúde Mental')
plt.legend(title='Sintoma')

plt.tight_layout()
plt.show()
```

#### 3.3.5 - A complexidade dos gêneros musicais tem alguma relação com algumas das condições mentais dos respondentes?

Neste trecho, investigamos a complexidade média das letras de músicas em relação a diferentes transtornos mentais, analisando como a complexidade varia com os níveis de ansiedade, depressão, TOC e insônia, agrupando os dados em intervalos de 0 a 10.

```python
complexidade_ansiedade = df_combinado.groupby(pd.cut(df['ansiedade'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Complexidade'].mean().reset_index()
complexidade_ansiedade['Transtorno'] = 'Ansiedade'

complexidade_depressao = df_combinado.groupby(pd.cut(df['depressao'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Complexidade'].mean().reset_index()
complexidade_depressao['Transtorno'] = 'Depressão'

complexidade_toc = df_combinado.groupby(pd.cut(df['toc'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Complexidade'].mean().reset_index()
complexidade_toc['Transtorno'] = 'TOC'

complexidade_insonia = df_combinado.groupby(pd.cut(df['insonia'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Complexidade'].mean().reset_index()
complexidade_insonia['Transtorno'] = 'Insônia'

complexidade_ansiedade.rename(columns={'ansiedade': 'Nível'}, inplace=True)
complexidade_depressao.rename(columns={'depressao': 'Nível'}, inplace=True)
complexidade_toc.rename(columns={'toc': 'Nível'}, inplace=True)
complexidade_insonia.rename(columns={'insonia': 'Nível'}, inplace=True)

df_combinado_complexidade = pd.concat([complexidade_ansiedade, complexidade_depressao, complexidade_toc, complexidade_insonia])

plt.figure(figsize=(12, 6))
sns.barplot(data=df_combinado_complexidade, x='Nível', y='Complexidade', hue='Transtorno')

plt.title('Complexidade Média das Letras por Nível de Transtornos')
plt.xlabel('Nível do Transtorno')
plt.ylabel('Complexidade Média')
plt.legend(title='Transtorno')
plt.show()
```

Observamos que entre os níveis 0-6, quanto maior o nível de TOC, mais os respondentes buscaram gêneros mais complexos, fato que continuou nos níveis entre 8-10.

#### 3.3.6 - A polaridade das letras dos gêneros musicais tem alguma relação com algumas das condições mentais dos respondentes?

Neste trecho, analisamos a diversidade lexical das letras de músicas em relação a diferentes transtornos mentais, avaliando a polaridade média com base nos níveis de ansiedade, depressão, TOC e insônia, agrupados em intervalos de 0 a 10.

```python
diversidade_ansiedade = df_combinado.groupby(pd.cut(df['ansiedade'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Polaridade'].mean().reset_index()
diversidade_ansiedade['Transtorno'] = 'Ansiedade'

diversidade_depressao = df_combinado.groupby(pd.cut(df['depressao'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Polaridade'].mean().reset_index()
diversidade_depressao['Transtorno'] = 'Depressão'

diversidade_toc = df_combinado.groupby(pd.cut(df['toc'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Polaridade'].mean().reset_index()
diversidade_toc['Transtorno'] = 'TOC'

diversidade_insonia = df_combinado.groupby(pd.cut(df['insonia'], bins=[0, 2, 4, 6, 8, 10]), observed=False)['Polaridade'].mean().reset_index()
diversidade_insonia['Transtorno'] = 'Insônia'

diversidade_ansiedade.rename(columns={'ansiedade': 'Nível'}, inplace=True)
diversidade_depressao.rename(columns={'depressao': 'Nível'}, inplace=True)
diversidade_toc.rename(columns={'toc': 'Nível'}, inplace=True)
diversidade_insonia.rename(columns={'insonia': 'Nível'}, inplace=True)

df_combinado_diversidade = pd.concat([diversidade_ansiedade, diversidade_depressao, diversidade_toc, diversidade_insonia])

plt.figure(figsize=(12, 6))
sns.barplot(data=df_combinado_diversidade, x='Nível', y='Polaridade', hue='Transtorno')

plt.title('Diversidade Lexical das Letras por Nível de Transtornos')
plt.xlabel('Nível do Transtorno')
plt.ylabel('Polaridade Média')
plt.legend(title='Transtorno')
plt.show()
```

Observamos aqui que, quanto maior o nível de ansiedade, mais os respondentes se interessaram por músicas com polaridades mais altas
